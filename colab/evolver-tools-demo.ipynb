### 1. 📊 CSV Analysis — Stats on the fly

Generate sample data and get instant stats, no pandas needed.

````python
import csv, io

# Generate sample sales data
data = io.StringIO()
writer = csv.writer(data)
writer.writerow(['product', 'price', 'units_sold', 'revenue', 'region'])
for i in range(20):
    writer.writerow([
        ['Widget A', 'Widget B', 'Gadget X', 'Gadget Y', 'Tool Z'][i % 5],
        [9.99, 14.99, 24.99, 49.99, 99.99][i % 5],
        [10, 25, 50, 100, 200][i % 5],
        [99.90, 374.75, 1249.50, 4999.00, 19998.00][i % 5],
        ['North', 'South', 'East', 'West'][i % 4]
    ])

with open('/tmp/sales.csv', 'w') as f:
    f.write(data.getvalue())

print("📄 Sales data generated (20 rows)")
!echo "---"
!evtool csv-stats /tmp/sales.csv
````

### 2. 📈 CSV Chart — Visualize in terminal

Bar charts right in your terminal. Great for quick data exploration.

````python
!evtool csv-chart /tmp/sales.csv --column revenue --title "Revenue by Product (sample)"
````

### 3. 🔐 Password Generator — Secure in one command

Generate cryptographically secure passwords of any length.

````python
!evtool gen-password 24
!evtool gen-password 32 --no-symbols
!evtool gen-password 16 --no-digits --no-symbols
````

### 4. 🎨 ASCII Art — Banners and more

Create ASCII banners for your projects, READMEs, or terminal welcome screens.

````python
!evtool ascii-banner "EVOLVER"
!echo "---"
!evtool figlet "Hello World"
````

### 5. 🖼 QR Code — Generate from text

Generate QR codes as PNG files. No external library needed.

````python
!evtool qrcode "https://evolver-dev.github.io/evolver-tools/" /tmp/evolver-qr.png
!echo "QR code saved to /tmp/evolver-qr.png"
!file /tmp/evolver-qr.png
````

### 6. 📋 JSON Processing — No jq needed

Pretty-print, validate, select, and transform JSON — all with zero dependencies.

````python
!echo '{"users":[{"name":"Alice","role":"engineer","skills":["python","go","k8s"]},{"name":"Bob","role":"designer","skills":["figma","css"]},{"name":"Charlie","role":"devops","skills":["aws","terraform","docker"]}]}' > /tmp/data.json
!evtool json-pretty /tmp/data.json
!echo "---"
!evtool json-validate /tmp/data.json
````

### 7. 💻 System Info — Quick diagnostics

CPU, memory, disk, and network info without third-party tools.

````python
!evtool sys-info
!echo "---"
!evtool disk-usage
````

### 8. 🔄 Unit Conversion — Everything in one command

Temperature, length, weight, volume, speed, and more.

````python
!evtool unit-convert 98.6 fahrenheit celsius
!evtool unit-convert 100 km mi
!evtool unit-convert 42 kg lb
!evtool unit-convert 2026-06-01 + 45 days
````

### 9. ⚙️ Text Processing — Swiss army knife

Base64 encode/decode, hash files, hex dumps, regex matching — all built-in.

````python
!echo "Hello from EVOLVER Tools!" | evtool base64
!echo "SGVsbG8gZnJvbSBFVk9MVkVSIFRvb2xzIQ==" | evtool base64 --decode
!echo "Hello from EVOLVER Tools!" > /tmp/message.txt && evtool hash-file /tmp/message.txt md5
````
