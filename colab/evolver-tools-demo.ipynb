## 1. Install

One command. That's it.

````python
!pip install -q evolver-tools
print('✅ evolver-tools installed')
!evtool --version 2>&1 | head -3
````

## 2. Browse Available Tools

See what's available — all 261 tools, organized by category.

````python
!evtool categories
````

````python
# Try: json, csv, network, crypto, ascii, sys, file, convert...
!evtool search csv
````

## 📊 Data & CSV Toolkit

EVOLVER has **20+ CSV tools** — no need for pandas for quick data tasks.

````python
# Create sample data
%%writefile sample.csv
name,role,score,department
Alice,Engineer,95,Data
Bob,Designer,88,Product
Charlie,PM,92,Product
Diana,Engineer,91,Data
Eve,Designer,85,Product
Frank,Engineer,97,Data
Grace,PM,89,Product
print('✅ sample.csv created')
````

````python
# Pretty-print as a table
!evtool csv-to-table sample.csv
````

````python
# Filter rows by column value
!evtool csv-filter sample.csv department Data
````

````python
# Full statistical analysis
!evtool csv-stats sample.csv --all
````

````python
# CSV → JSON conversion
!evtool csv2json sample.csv --pretty
````

---
## { } JSON Toolkit

**18+ JSON tools** — validate, query, diff, merge, flatten, schema inference.
No more `jq` installation headaches.

````python
# Pretty-print JSON
!echo '{"project":"evolver-tools","version":"38.0","tools":261,"categories":{"csv":20,"json":18,"network":22,"security":14,"ascii":12}}' | evtool json-pretty
````

````python
# Query nested JSON with dot-path syntax
!echo '{"data":{"users":[{"name":"Alice","age":30},{"name":"Bob","age":25}]}}' | evtool json-path data.users
````

````python
# SQL-style queries on JSON arrays
!echo '[{"name":"Alice","role":"engineer"},{"name":"Bob","role":"designer"}]' | evtool jsonql "SELECT * FROM ? WHERE role='engineer'"
````

## 🔗 Pipe Workflows

The real power: chain tools together with Unix pipes.

````python
# CSV → JSON → query → pretty-print
!cat sample.csv | evtool csv2json --pretty | evtool json-pretty | head -20
````

````python
# Data → chart → banner
!echo -e 'Python:78\nJavaScript:65\nRust:42\nGo:38' | evtool chart-cli bar
````

````python
# Hash a file → encode → display
!head -3 sample.csv | evtool b64
````
